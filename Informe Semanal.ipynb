# Años Individuales

###### 2022

```python

Negocios1 = Negocios.drop(Negocios[Negocios['Logro: Eje'] != 'Exportaciones'].index)
Negocios1 = Negocios1.drop(Negocios1[Negocios1['Meta 1 responsable Colombia: Año'] !=  añoactual].index)
Negocios1 = Negocios1.drop(Negocios1[Negocios1['Logro: Tipo de registro'] !=  'Monto de Exportación'].index)

#Monto
monto2022 = round((Negocios1['Logro: Valor'].sum())/1000000, 2)
F1[fechaactual] = 0
F1.loc[F1['Indicador'] == 'Negocios Exportaciones US$ millones', fechaactual] = monto2022
#Empresas Nacionales
EN = Negocios1.groupby(['Logro: ID Cuenta nacional' , 'Logro: Cuenta nacional' ])['Logro: Valor'].sum().reset_index()
ENacio = len(set(EN['Logro: ID Cuenta nacional']))
F1.loc[F1['Indicador'] == 'Empresas Col con negocios', fechaactual] = ENacio
#Compradores con negocios
CN = Negocios1.groupby(['Logro: ID Cuenta exterior' , 'Logro: Cuenta exterior' ])['Logro: Valor'].sum().reset_index()
CN = len(set(CN['Logro: ID Cuenta exterior']))
F1.loc[F1['Indicador'] == 'Compradores con negocios', fechaactual] = CN
#Departamentos
Depto = Negocios1.groupby(['C_Nal_Departamento' ])['Logro: Valor'].sum().reset_index()
Depto = len(set(Depto['C_Nal_Departamento']))
F1.loc[F1['Indicador'] == '# Departamentos', fechaactual] = Depto
#Mercado
Pais = Negocios1.groupby(['Logro: Mercado' ])['Logro: Valor'].sum().reset_index()
Pais = len(set(Pais['Logro: Mercado']))
F1.loc[F1['Indicador'] == 'Países de los negocios', fechaactual] = Pais
#No. MyPimes
MyP = Negocios1.groupby(['Nuevo Tamaño', 'Logro: ID Cuenta nacional' , 'Logro: Cuenta nacional'])['Logro: Valor'].sum().reset_index()
MyP = MyP.drop(MyP[(MyP['Nuevo Tamaño'] !=  'Pequeña') &    
                                     (MyP['Nuevo Tamaño'] !=  'Mediana') & 
                                     (MyP['Nuevo Tamaño'] !=  'Microempresa')].index)
MyP = len(set(MyP['Logro: ID Cuenta nacional']))
F1.loc[F1['Indicador'] == 'No. Mipymes', fechaactual] = MyP

#No. Medianas
Mediana = Negocios1.groupby(['Nuevo Tamaño', 'Logro: ID Cuenta nacional' , 'Logro: Cuenta nacional'])['Logro: Valor'].sum().reset_index()
Mediana = Mediana.drop(Mediana[(Mediana['Nuevo Tamaño'] !=  'Mediana')].index)
Mediana = len(set(Mediana['Logro: ID Cuenta nacional']))
F1.loc[F1['Indicador'] == 'No. Medianas', fechaactual] = Mediana

#Acuerdo
Negocios2 = Negocios1.drop(Negocios1[Negocios1['E_AcuerdoComercialEXP'] == 'Z_Mercado no cuenta con Acuerdo Comercial Vigente'].index)
#Monto
montoa2022 = round((Negocios2['Logro: Valor'].sum())/1000000, 2)
F1.loc[F1['Indicador'] == 'Monto de negocios a mercados con acuerdo', fechaactual] = montoa2022

#% negocios a mercados con acuerdo
porac =  round((montoa2022/monto2022)*100,0)
F1.loc[F1['Indicador'] == '% negocios a mercados con acuerdo', fechaactual] = porac

#Empresas Nacionales
EN = Negocios2.groupby(['Logro: ID Cuenta nacional' , 'Logro: Cuenta nacional' ])['Logro: Valor'].sum().reset_index()
ENacio = len(set(EN['Logro: ID Cuenta nacional']))
F1.loc[F1['Indicador'] == 'Empresas Col con negocios en mercados con Acuerdo', fechaactual] = ENacio

#Naranja
Negocios3 = Negocios1.drop(Negocios1[Negocios1['E_Economía_Naranja'] != 1].index)
#Empresas Nacionales
EN = Negocios3.groupby(['Logro: ID Cuenta nacional' , 'Logro: Cuenta nacional' ])['Logro: Valor'].sum().reset_index()
ENacio = len(set(EN['Logro: ID Cuenta nacional']))
F1.loc[F1['Indicador'] == 'Empresas Col con negocios (Economía Naranja)', fechaactual] = ENacio
#Compradores con negocios
CN = Negocios3.groupby(['Logro: ID Cuenta exterior' , 'Logro: Cuenta exterior' ])['Logro: Valor'].sum().reset_index()
CN = len(set(CN['Logro: ID Cuenta exterior']))
F1.loc[F1['Indicador'] == 'Compradores con negocios (Economía Naranja)', fechaactual] = CN
#Monto
montoa2022 = round((Negocios3['Logro: Valor'].sum())/1000000, 2)
F1.loc[F1['Indicador'] == 'Negocios Exportaciones US$ millones (Economía Naranja)', fechaactual] = montoa2022
#Departamentos
Depto = Negocios3.groupby(['C_Nal_Departamento' ])['Logro: Valor'].sum().reset_index()
Depto = len(set(Depto['C_Nal_Departamento']))
F1.loc[F1['Indicador'] == '# Departamentos (Economía Naranja)', fechaactual] = Depto
#Mercado
Pais = Negocios3.groupby(['Logro: Mercado' ])['Logro: Valor'].sum().reset_index()
Pais = len(set(Pais['Logro: Mercado']))
F1.loc[F1['Indicador'] == 'Países de los negocios (Economía Naranja)', fechaactual] = Pais

#Servicios
Negocios4 = Negocios1.drop(Negocios1[Negocios1['E_Bienes'] == 1.0].index)
#Empresas Col
EN = Negocios4.groupby(['Logro: ID Cuenta nacional' , 'Logro: Cuenta nacional' ])['Logro: Valor'].sum().reset_index()
ENacio = len(set(EN['Logro: ID Cuenta nacional']))
F1.loc[F1['Indicador'] == 'Empresas Col con negocios (Servicios)', fechaactual] = ENacio
#Compradores con negocios
CN = Negocios4.groupby(['Logro: ID Cuenta exterior' , 'Logro: Cuenta exterior' ])['Logro: Valor'].sum().reset_index()
CN = len(set(CN['Logro: ID Cuenta exterior']))
F1.loc[F1['Indicador'] == 'Compradores con negocios (Servicios)', fechaactual] = CN
#Monto
montoa2022 = round((Negocios4['Logro: Valor'].sum())/1000000, 2)
F1.loc[F1['Indicador'] == 'Negocios Exportaciones US$ millones (Servicios)', fechaactual] = montoa2022
#Departamentos
Depto = Negocios4.groupby(['C_Nal_Departamento' ])['Logro: Valor'].sum().reset_index()
Depto = len(set(Depto['C_Nal_Departamento']))
F1.loc[F1['Indicador'] == '# Departamentos (Servicios)', fechaactual] = Depto
#Mercado
Pais = Negocios4.groupby(['Logro: Mercado' ])['Logro: Valor'].sum().reset_index()
Pais = len(set(Pais['Logro: Mercado']))
F1.loc[F1['Indicador'] == 'Países de los negocios (Servicios)', fechaactual] = Pais
```
