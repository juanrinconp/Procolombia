```python
MetasH = Metas.drop(Metas[Metas['Meta: Eje'] != 'Inversión'].index)

MetasH = MetasH.drop(MetasH[MetasH['Meta: Año'] !=  añoactual].index)

MetasH = MetasH.drop(MetasH[MetasH['Meta: Tipo meta funcionario'] != 'EXTERIOR'].index)



MetasQH = MetasH.drop(MetasH[MetasH['Meta: Período'] > trimestre].index)
```

```python
aH = MetasH.groupby(['E_Localización' ,'Meta: Nombre del propietario' ,'Meta: Tipo de meta' ])['Meta: Valor meta'].sum().reset_index()
a2H = MetasQH.groupby(['E_Localización' ,'Meta: Nombre del propietario' ,'Meta: Tipo de meta'])['Meta: Valor meta'].sum().reset_index()

MetasdfH = aH.merge(a2H, how='outer', left_on=['E_Localización' ,  'Meta: Tipo de meta' , 'Meta: Nombre del propietario'], right_on=['E_Localización' , 'Meta: Tipo de meta' , 'Meta: Nombre del propietario']).fillna(0)
MetasdfH.rename(columns={'Meta: Valor meta_x': 'Valor Meta', 'Meta: Valor meta_y': 'Valor Meta Q'}, inplace=True)
```

```python
NegociosH = Negocios.drop(Negocios[Negocios['Logro: Eje'] != 'Inversión'].index)
NegociosH = NegociosH.drop(NegociosH[NegociosH['Meta 1 responsable Colombia: Año'] !=  añoactual].index)
NegociosH = NegociosH.drop(NegociosH[NegociosH['Logro: Tipo de registro'] !=  'Monto de Inversión'].index)
```
